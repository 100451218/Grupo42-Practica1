```python
import pandas as pd
#Importamos los datos.
comp_df = pd.read_csv("comp_st42ns1.txt.bz2", 
 compression="bz2", 
 index_col=0)
#print(disp_df['apcp_sf1_1'][0])
pd.set_option('display.max_columns', None)
comp_df = comp_df.drop(columns=['uswrf_s1_1'])
display(comp_df)
```
